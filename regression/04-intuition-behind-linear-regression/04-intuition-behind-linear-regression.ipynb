# Intuition behind linear regression

### Linear function

```python
def f(x):
    '''Linear function to calculate the value of the dependent variable y (outcome).'''
    w = -2  # weight
    b = 8 # bias
    return w*x + b


# changing the value of the independent variable x (predictor):
for x in range(0, 6):
    print(f"y = f({x}) =", f(x))
```

### Plotting a line

```python
import matplotlib.pyplot as plt
import numpy as np

plt.xlabel("x", rotation=0, labelpad=10)
plt.ylabel("y", rotation=0, labelpad=15)

arr = np.arange(0, 10)

plt.plot(arr, f(arr), "r")

plt.show()
```
